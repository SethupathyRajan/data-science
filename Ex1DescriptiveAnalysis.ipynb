```python
import kagglehub
import pandas as pd
from kagglehub import KaggleDatasetAdapter

#Load Dataset
file_path = "data.csv"

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "shriyashjagtap/indian-personal-finance-and-spending-habits",
    file_path,
)

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#Selecting numerical datatypes
num_df = df.select_dtypes(include=['int64', 'float64'])
num_df = num_df.drop(columns=['Age'])
num_df.head()
```

1. Measure of Central Dependency

```python
#1.Mean

mean_value = num_df.mean().round()
print(mean_value)
```

```python
#Median
median_value = num_df.median().round()
print(median_value)
```

```python
print(len(num_df[num_df['Loan_Repayment']==0]))
```

```python
#Mode
mode_value = num_df['Loan_Repayment'].mode()
print(mode_value)
```

3. Measure of Dispersion

```python
range_val = num_df.max() - num_df.min()
range_val
```
