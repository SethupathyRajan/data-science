```python
import kagglehub
import pandas as pd
from kagglehub import KaggleDatasetAdapter

#Load Dataset
file_path = "data.csv"

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "shriyashjagtap/indian-personal-finance-and-spending-habits",
    file_path,
)

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#Selecting numerical datatypes
num_df = df.select_dtypes(include=['int64', 'float64'])
num_df = num_df.drop(columns=['Age'])
num_df.head()
```

1. Measure of Central Dependency

```python
#1.Mean

mean_value = num_df.mean().round()
print(mean_value)
```

```python
#Median
median_value = num_df.median().round()
print(median_value)
```

```python
print(len(num_df[num_df['Loan_Repayment']==0]))
```

```python
#Mode
mode_value = num_df.mode(axis=0).iloc[0]
print(mode_value)
```

3. Measure of Dispersion

```python
range_val = num_df.max() - num_df.min()
range_val
```

```python
#variance

var =num_df.var()
var
```

```python
#Standard Deviation

std_value = num_df.std()
std_value
```

```python
#IQR

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)

IQR = Q3 - Q1
IQR
```

Shape of Distribution


```python
skewness = num_df.skew()
skewness
```

```python
#kurtosis

kurtosis = num_df.kurtosis()
kurtosis
```
