```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_name = "data.csv"

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "shriyashjagtap/indian-personal-finance-and-spending-habits",
    file_name,
)

df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

Defining the Target Variable


```python
spending = np.array(df['Disposable_Income'])
spending

smin = min(spending)
smax = max(spending)

print(smin)
print(smax)

np.linspace(smin,smax,4)
```

```python
df['Spending_Class'] = pd.cut(
    df['Disposable_Income'],
    bins=[ -5400.78867337, 122086.21371166, 249573.21609668, 377060.2184817 ],
    labels=['Overspender','Moderate','Saver']
)
```

```python
df['Spending_Class'].value_counts()
```

```python
df.head()
```

1 hot Encoding


```python
df = pd.get_dummies(df, columns=['Occupation', 'City_Tier'], drop_first=True)
```

Define target values

```python
X = df.drop(['Spending_Class','Disposable_Income'],axis=1)
Y = df['Spending_Class']
```

Train test split

```python
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test = train_test_split(
    X,Y,test_size=0.20,random_state=42
)

```

Feature Scaling for KNN & logistic Regression

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```


Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(x_train,y_train)

lr_pred = lr_model.predict(x_test)
```

KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train,y_train)

knn_pred = knn_model.predict(x_test)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train,y_train)

dt_pred = dt_model.predict(x_test)
```

Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train,y_train)

rf_pred = rf_model.predict(x_test)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score , classification_report

print(classification_report(y_test, rf_pred))
print(classification_report(y_test, dt_pred))
print(classification_report(y_test, knn_pred))
print(classification_report(y_test, lr_pred))
```

```python
# Diagnostic cell: check for perfect single-feature predictors and leakage
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

print('X shape:', X.shape)
print('Y distribution:\n', Y.value_counts())

# Constant columns
const_cols = [c for c in X.columns if X[c].nunique() == 1]
print('Constant columns:', const_cols)

# If x_train/x_test are numpy arrays, recreate a split from X/Y for diagnostics
try:
    X_df = pd.DataFrame(x_train, columns=X.columns) if not isinstance(x_train, pd.DataFrame) else x_train
    xtr, xts, ytr, yts = x_train, x_test, y_train, y_test
except Exception:
    xtr, xts, ytr, yts = train_test_split(X, Y, test_size=0.20, random_state=42)

perfect_single = []
for c in X.columns:
    clf = DecisionTreeClassifier(random_state=42)
    # ensure 2D input
    clf.fit(xtr[[c]], ytr)
    acc = accuracy_score(yts, clf.predict(xts[[c]]))
    print(f"{c}: {acc:.4f}")
    if acc == 1.0:
        perfect_single.append(c)

print('Perfect single-feature predictors:', perfect_single)

# Check correlations with Disposable_Income if present
if 'Disposable_Income' in df.columns:
    print('\nTop absolute correlations with Disposable_Income:')
    print(df.corr()['Disposable_Income'].abs().sort_values(ascending=False).head(10))
```
